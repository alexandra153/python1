para poder usar dataframe vamos a usar dos librerias una es nompy y la otra es pandas las dos se complementan y hacen cosas diferentes
Para crear data frame existen varias posibilidades, sin embargo la más usada es la lectura directa desde un archivo.

```python
#Opción 1 --> se agrega un alias para no llamar toda la libreria usamos para esto pd y np
import pandas as pd
import numpy as np
```

```python
#la primera forma de crear un dataframe es usando el metodo array y luego guradarlo en una variable
dataframe_1=np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#para crear el dataframe del arreglo anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2=np.array([[5,10],[15,20],[25,30],[35,40],[45,50]])
```

```python
pd.DataFrame(dataframe_2)
```

```python
dataframe_3=np.array([[3,6,9],[12,15,18],[21,24,27],[30,0,0]])
```

```python
pd.DataFrame(dataframe_3)
```

```python
#ahora vamos a cambiar el nombre de las columnas
df1=pd.DataFrame(dataframe_1, index=["fila1", "fila2", "fila3","fila4"],
                         columns=["columna1", "columna2"])
```

```python
df1
```

```python
#opción 2
#Ahora creamos un dataframe usando listas
dataframe2=[[2,4],[6,8],[10,12]]
```
